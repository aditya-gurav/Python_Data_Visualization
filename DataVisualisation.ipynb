Advanced Visualizations - Exploratory Data Analysis (EDA)

1. Scatter Plot
2. Relational Plot
3. Box Plot
4. Pair Plot
5. Heatmap
6. Cat Plot

```python
import pandas as pd #pd is the alias of pandas
```

```python
data=pd.read_csv("/content/tips.csv")
```

```python
data.head()
```

```python
data.plot.scatter(x='total_bill',y='tip', color='r') #scatter plot
```

```python
snr.relplot(x='total_bill',y='tip',data=data) #relational plot
```

```python
snr.relplot(x='total_bill',y='tip',hue='sex', style='smoker',col='day', size='size', data=data)
#hue is a color parameter - any color we put inside hue will become colourfull
#style is a parameter which gives shapes to the categories - any coloumn we put inside style, the categories inside will get a shape
#col is a parameter which will give number of plots same as number of categories in the coloumn
```

Q. Visualize the relation betn total bill collected and tip for lunch and dinner time using relational plot

```python
snr.relplot(x='total_bill',y='tip',col='time', data=data)
```

```python
snr.pairplot(data) #pair_plot
```

```python
snr.catplot(x='day',y='total_bill', hue='sex', data=data) #cat_plot
```

```python
snr.catplot(x='day',y='total_bill', hue='sex', kind='bar', data=data)
```

```python
snr.catplot(x='day',y='total_bill', hue='sex', kind='point', data=data)
```

```python
snr.catplot(x='day',y='total_bill', hue='sex', kind='violin', data=data)
```

```python
snr.catplot(x='day',y='total_bill', hue='sex', kind='box', data=data)
```
